```
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.image as mpimg
```

```
image = mpimg.imread('lena.png')
plt.grid(False)
plt.imshow(image)
```

```
# Converting it into the black and white image

# To convert it into b&w take the mean of the 3color channel

bw = np.mean(image, axis=2)  # axis=2 is the color channel
plt.grid(False)
plt.imshow(bw, cmap='gray')
```

```
# Creating the Gaussian filter

gf = np.zeros((20,20))

for i in range(20):
  for j in range(20):
    dis = (i-9.5)**2 + (j-9.5)**2
    gf[i,j] = np.exp(-dis / 50.)  # taking the -ive value divde by a big number
    
gf = gf / gf.sum()
```

```
plt.grid(False)
plt.imshow(gf, cmap='gray')
```

```
# Convolve the image with Gaussian filter
out = convolve2d(bw, gf)
plt.grid(False)
plt.imshow(out, cmap='gray')
```

```
print('size of original image: ',image.shape)
print('size of b&w image: ',out.shape)
```

```
# what's that weird black stuff on the edges? let's check the size of output

# after convolution, the output signal is N1 + N2 - 1

# The size of the image increased
```

```
# we can also just make the output the same size as the input
out = convolve2d(bw, gf, mode='same')
plt.imshow(out, cmap='gray')
plt.grid(False)
plt.show()
```
